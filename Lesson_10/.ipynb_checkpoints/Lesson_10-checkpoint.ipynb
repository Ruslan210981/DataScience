# Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
def det_class(row):
    if row['rating'] <= 2.0: return 'низкий рейтинг'
    if row['rating'] <= 4.0: return 'средний рейтинг'
    if row['rating'] <= 5.0: return 'высокий рейтинг'

df1 = pd.read_csv('ratings.csv')
df1 = df1.groupby(['movieId'])['rating'].mean().reset_index()
df1['class'] = df1.apply(det_class, axis = 1)
df1

```

# Задание 2

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

```python
# %%time 
# универсальный метод
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}

def geo_class(req):
    req = req.split(' ')
    for word in req:
        for key, value in geo_data.items():
            if word in value: return key
    return 'undefined'

df1 = pd.read_csv('keywords.csv')
df1['region'] = df1['keyword'].apply(geo_class)
df1.head(10)




```

```python
# %%time
# частный случай, но немного побыстрее
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}

def geo_class(req):
    req = req.split(' ')
    for word in req:
        if word in geo_data['Центр']: return 'Центр'
        if word in geo_data['Северо-Запад']: return 'Северо-Запад'
        if word in geo_data['Дальний Восток']: return 'Дальний Восток'
    return 'undefined'

df1 = pd.read_csv('keywords.csv')
df1['region'] = df1['keyword'].apply(geo_class)
df1.head(10)
```
