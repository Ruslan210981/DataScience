# Задание 1. Загрузка данных
Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
import pandas as pd
horse_df = pd.DataFrame()
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

def decipher(row):
    line = row.strip().split(',')
    for i in range(0, len(line)):
        if line[i] == '?' : 
            line[i] = 4
        else: 
            try:
                line[i] = int(line[i])
            except:
                line[i] = float(line[i])
    return line
            

horse_data = open('Horse_data.csv','r')
for line in horse_data:
    line = decipher(line)
    row = {'surgery' : line[0], 'Age' : line[1], 'rectal temperature' : line[3], 'pulse' : line[4],
          'respiratory rate' : line[5], 'temperature of extremities' :line[6], 'pain' :line[10],
          'outcome' : line[22], 'type of lesion1' : line[24], 'type of lesion2' : line[25], 'type of lesion3' :line[26],
          'abdominal distension' : line[12], 'nasogastric tube' : line[13], 'nasogastric reflux' :line[14], 'nasogastric reflux PH' : line[15],
          'rectal examination - feces' : line[16], 'abdomen' : line[17], 'packed cell volume' :line[18], 'total protein' : line[19],
          'abdominocentesis appearance' : line[20], 'abdomcentesis total protein' : line[21], 'surgical lesion' : line[23],
          'mucous membranes' : line[8], 'capillary refill time' : line[9], 'cp_data' : line[27], 
            'Hospital Number' : line[2], 'peripheral pulse' : line[7], 'peristalsis' : line[11],
          }
    horse_df = pd.concat([horse_df, pd.DataFrame([row])])
# horse_df['respiratory rate'].fillna(10, inplace = True)
horse_df.groupby(['outcome','type of lesion1'])['Age'].count()
# horse_df
```
