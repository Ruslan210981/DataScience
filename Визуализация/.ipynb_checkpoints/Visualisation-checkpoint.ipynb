```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 20,10
app_df = pd.DataFrame()
app_data = open('googleplaystore.csv', 'r', encoding='utf-8')
fields = next(app_data)
fields = fields.split(',')
for line in app_data: 
    #  - избавляемся от запятых в кавычках
    line = line.split('"')
    res_line = ''
    for i in range(len(line)):   # в каждом нечетном слове, т.е. в слове обрамленном кавычками убираем запятую
        if i%2 : line[i] = line[i].replace(',','').replace('+','')
        line[i] = line[i].replace('+','')
        res_line += line[i]      # собираем обновленную строчку в единое слово для дальнейшего разбиения
        # - преобразуем в список 
    res_line = res_line.split(',')
    if len(res_line) < 13: 
        continue
    row = {fields[0] : res_line[0], fields[1] : res_line[1], fields[2] : float(res_line[2]), fields[3] : int(res_line[3]), fields[4] : res_line[4],
           fields[5] : int(res_line[5]), fields[6] : res_line[6], fields[7] : res_line[7], fields[8] : res_line[8],
           fields[9] : res_line[9], fields[10] : res_line[10], fields[11] : res_line[11], fields[12] : res_line[12]}
    app_df = pd.concat([app_df, pd.DataFrame([row])])
app_df.head()
```

```python
# определяем самое большое приложение 
# удаляем буквы из размера и преобразуем все в кБ.
app_df['Size'] = app_df['Size'].apply(lambda x: str(x).replace('M','000'))
app_df['Size'] = app_df['Size'].apply(lambda x: str(x).replace('k',''))
app_df['Size'] = app_df['Size'].apply(lambda x: str(x).replace('.',''))
app_df['Size'] = pd.to_numeric(app_df['Size'],errors='coerce')
print('Приложение с самым большим размером -', app_df.sort_values('Size', ascending=False).iloc[0]['App'])
```

# Строим график зависимости установок от категории приложения

```python
stat_Installs = app_df[['Installs','Category']].groupby('Category').sum()
stat_Installs.plot(kind = 'bar', title = 'Зависимость установок от категории',
                      xlabel = 'Category', ylabel = 'Кол-во установок')
```

По графику можно сделать вывод, что самыми популярными категориями считаются игры и мессенджеры

```python
stat_Apps = app_df[['App','Category']].groupby('Category').count()
stat_Apps.plot(kind = 'bar', title = 'Количество приложений в категориях',
                      xlabel = 'Category', ylabel = 'Кол-во приложений в категории')
```

По графику видим что больше всего приложений в категориях Family, Game, Tools.

# Построим график соотношения платных и бесплатных приложений.

```python
stat_Type = app_df[['Installs','Type']].groupby('Type').count()
stat_Type.plot(kind = 'bar', title = 'Соотношение платных приложений к бесплатным',
                      xlabel = 'Тип приложения', ylabel = 'Кол-во установок')
```

По графику видно что бесплатных приложений значительно больше платных

```python
app_df.groupby('Content Rating')['Installs'].count().plot(kind = 'bar', title = 'Зависимость установок от категории',
                      xlabel = 'Category', ylabel = 'Кол-во установок')
```

Можем сделать вывод что приложений без возрастных ограничений больше всего.
